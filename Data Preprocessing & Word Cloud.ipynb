```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from sklearn.feature_extraction.text import TfidfVectorizer
```

#### stopwords

```python
#english = pd.read_csv('english.txt', header=None)
#spanish = pd.read_csv('spanish.txt', header=None)
```

```python
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('spanish') + stopwords.words('english') + ["could"]
```

```python
tracks_lyric = pd.read_csv('spotify_tracks.csv', index_col=0)
corpus = tracks_lyric.lyrics
```

```python
top10_id = ['25sgk305KZfyuqVBQIahim','0rTV5WefWd1J3OwIheTzxM','1rqqCSm0Qe4I9rUvWncaom',
            '116H0KvKr2Zl4RPuVBruDO','7COXchtUOMd6uIT6HvmRaI','2IRZnDFmlqMuOrYOLnZZyc','5itOtNx0WxtJmi1TQ3RuRd',
           '2xLMifQCjDGFmkHkpNLD9h','7eBqSVxrzQZtK2mmgRG6lC','0FZ4Dmg8jJJAPJnvBIzD9z']

# Creating corpus for top 10 English words
top_lyrics = pd.Series()

for track_id in top10_id:
    temp = tracks_lyric.loc[tracks_lyric['id'] == track_id,'lyrics']
    top_lyrics = top_lyrics.append(pd.Series(temp))
```

```python
all_lyrics = ', '.join(tracks_lyric.lyrics)
corpus = corpus.append(pd.Series(all_lyrics))

top_lyrics = ', '.join(top_lyrics)
corpus = corpus.append(pd.Series(top_lyrics))
```

```python
np.size(tracks_lyric['id']),np.size(corpus)
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords, ngram_range = (1,1), max_df = .6, min_df = .01)
```

```python
# Apply TfidfVectorizer to speech texts. X is a sparse matrix which holds the location of all non-zero values.
X = vectorizer.fit_transform(corpus)
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
feature_names = vectorizer.get_feature_names()
dense = X.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
df.head()
```

```python
# Prep TF-IDF Matrix for Word Clouds
names = tracks_lyric.id.append(pd.Series('all_lyrics'))
names = names.append(pd.Series('top10_lyrics'))
data = df.transpose()
data.columns = names
data.head()
```

### wordcloud for all tracks in total

```python
# wordcloud for all songs
wc_all_lyrics = WordCloud(width = 1920, height = 1080).generate_from_frequencies(data['all_lyrics'])
plt.figure(figsize=(15,8))
plt.imshow(wc_all_lyrics, interpolation='bilinear')
plt.axis("off")
plt.show()
```
